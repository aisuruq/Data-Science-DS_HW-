## Что такое токенизатор?

**Токенизатор** — это инструмент или процесс, который разбивает текст на отдельные элементы, называемые токенами. Эти токены могут быть словами, частями слов или даже символами. Это важный шаг в обработке текста, который помогает программам понять структуру языка и работать с текстом эффективно.

- Надо смотреть подходит ли он под вашу задачу или нет, они все разные
- Если на хакатоне нужен ML, то надо сразу же сходит на hugging face

`моделька и обучающие данные мы ее на этих данных обучаем - обновим ее состояние она станет предсказывать то что нам нужно, а потом пользовательский запрос будет отправляться в это дообученную модельку и ее ответ будет нашим выходом`

*Модели должны поддерживать дообучение (fine-tuning), что позволяет адаптировать их к новым данным, если ты захочешь адаптировать модель под собственные источники данных*



### Команда	train_df
1. **train_df.head()**
- Показывает первые 5 строк датафрейма train_df.
- Используется для быстрого просмотра структуры данных.
- Можно задать количество строк, например: train_df.head(10).
2. **print(train_df.shape)**
- Выводит размерность датафрейма в формате (число строк, число столбцов).
- Например, если train_df.shape вернет (1000, 10), значит, у вас 1000 строк и 10 столбцов.

- train_df.tail(5) | Показывает последние 5 строк датафрейма
- train_df.info() | Выводит информацию о столбцах, типах данных и пропущенных значениях
- train_df.describe() | Рассчитывает базовую статистику (среднее, стандартное отклонение, минимум, максимум)
- train_df.columns | Показывает названия столбцов
- train_df.dtypes | Выводит типы данных всех столбцов
- train_df.isnull().sum() | Показывает количество пропущенных значений в каждом столбце
- train_df.nunique() | Показывает количество уникальных значений в каждом столбце
- train_df['column_name'].value_counts() | Выводит количество уникальных значений в указанном столбце
- train_df.sample(5) | Показывает случайные 5 строк из датафрейма
- train_df.corr() | Считает корреляцию между числовыми столбцами
- train_df.sort_values(by='column_name', ascending=False) | Сортирует датафрейм по столбцу
- train_df.drop(columns=['column1', 'column2'], inplace=True) | Удаляет указанные столбцы

### Команда val_df 
**val_df** — это, скорее всего, валидационный датафрейм (validation dataset), который используется для проверки качества модели на данных, не участвующих в обучении

##### Отличие train_df и val_df:
- train_df → Данные, на которых модель обучается.
- val_df → Данные, на которых модель проверяется во время обучения (но не используется для финального тестирования).
**Аналогичные команды для val_df:**
- val_df.head() → первые 5 строк валидационного набора.
- print(val_df.shape) → размерность val_df, например, (200, 10) (значит, 200 строк и 10 столбцов).
- val_df.info() → информация о типах данных.
- val_df.describe() → статистика по числовым столбцам.
- val_df.isnull().sum() → проверка пропущенных значений


`По данным правительства India AQI (CPCB), диапазон в Индии составляет от 0 до 500, где 0 — хорошо, а 500 — тяжело. При расчете AQI необходимо учитывать восемь основных загрязнителей, а именно. твердые частицы (PM10 и PM2.5), окись углерода (CO), озон (O3), двуокись азота (NO2), двуокись серы (SO2), аммиак (NH3) и свинец (Pb). Для расчета AQI должны присутствовать данные как минимум о трех загрязняющих веществах, одним из которых должен быть PM10 или PM2.5. AQI в диапазоне от 0 до 500 имеет разные концентрации для каждого загрязнителя и соответственно влияет на здоровье.`
